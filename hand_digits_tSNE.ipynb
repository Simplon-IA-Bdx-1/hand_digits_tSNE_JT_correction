```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import pylab
import tsne
import seaborn as sns
```

```python
X = np.load('data/X.npy')
print(X.shape)
Y = np.load('data/Y.npy')
```

```python
X_embedded = tsne.tsne(X.reshape((X.shape[0],X.shape[1]*X.shape[2])), 2, 50, 20.0)
```

```python
print(X_embedded.shape)
print(type(X_embedded))
X_df = pd.DataFrame(X_embedded)
Y_class = np.argmax(Y,axis=1)
Y_class = Y_class.reshape((Y_class.shape[0],1))

#print(Y_class.tostring())
X_df['class'] = Y_class
X_df.columns
X_df['class'] = 'classe '+ X_df['class'].astype(str)
X_df['class'].head()
```

```python
sns.set(rc={'figure.figsize':(12,10)})
ax = sns.scatterplot(x=X_df[0], y=X_df[1], data=X_df, hue='class')
```
